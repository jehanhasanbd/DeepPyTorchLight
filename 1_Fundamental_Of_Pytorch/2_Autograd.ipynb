# Perceptron Backpropagation

```python
x = torch.tensor(6.7)
y_true = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
print(x.shape)
print(x.unsqueeze(dim=0).shape)
```

```python
z = torch.dot( x.unsqueeze(0) , w.unsqueeze(0) ) + b.unsqueeze(0)
y_pred = torch.sigmoid(z)
y_pred
```

```python
def binary_cross_entropy(y_true, y_pred):
    ephcilon = 1e-8
    y_pred = torch.clamp(y_pred,ephcilon, 1-ephcilon)
    return -(y_true * torch.log(y_pred) + (1-y_true)*torch.log(1-y_pred))
```

```python
binary_cross_entropy(torch.tensor(1.0), torch.tensor(0.68))
```

```python
loss = binary_cross_entropy(y_true, y_pred)
loss
```

```python
loss.backward()
print(w.grad, b.grad)
```

# Multi Layer Perceptron

```python
X = torch.tensor([3.9,8.9 ,90])
y_true = torch.tensor(0.0)
```

```python
W1 = torch.rand(3,3, requires_grad=True)    
B1 = torch.rand(3, requires_grad=True)
print(W1,"\n",B1)
```

```python
# X(1,3) * W1(3,3) = (1,3) + B1(1,3)  = z1(1,3)
z1 = torch.matmul(X, W1) + B1
z1
```

```python
o1 = torch.relu(z1)
o1
```

```python
# X(1,3) * W2(3,3) = (1,3) + B2(1,3)  = z2(1,3)
W2 = torch.rand(3,3, requires_grad=True)    
B2 = torch.rand(3, requires_grad=True)
z2 = torch.matmul(o1, W2) + B2
o2 = torch.sigmoid(z2)
```

```python
loss = binary_cross_entropy(y_true, o2)
loss
```

```python
loss.backward()
```

```python
W1.grad
```
