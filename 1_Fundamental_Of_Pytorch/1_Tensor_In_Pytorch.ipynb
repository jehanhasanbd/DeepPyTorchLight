# iv) Statistical Operation
#### (Col wise--> 0) (Row wise --> 1)

```python
x
```

```python
print(torch.sum(x))
print("------------")
print(torch.sum(x, dim=0))    # col wise
print("------------")
print(torch.sum(x, dim=1))    # row wise
```

```python
print(torch.mean(x))
print("------------")
print(torch.mean(x, dim=0))    # col wise
print("------------")
print(torch.mean(x, dim=1))    # row wise
```

```python
print(torch.min(x))
print("------------")
print(torch.min(x, dim=0))    # col wise
print("------------")
print(torch.min(x, dim=1))    # row wise
```

```python
print(torch.max(x))
print("------------")
print(torch.max(x, dim=0))    # col wise
print("------------")
print(torch.max(x, dim=1))    # row wise
```

```python
print(torch.var(x))
print("------------")
print(torch.var(x, dim=0))    # col wise
print("------------")
print(torch.var(x, dim=1))    # row wise
```

```python
print(torch.argmin(x))
print("------------")
print(torch.argmin(x, dim=0))    # col wise
print("------------")
print(torch.argmin(x, dim=1))    # row wise
```

```python
print(torch.argmax(x))
print("------------")
print(torch.argmax(x, dim=0))    # col wise
print("------------")
print(torch.argmax(x, dim=1))    # row wise
```
