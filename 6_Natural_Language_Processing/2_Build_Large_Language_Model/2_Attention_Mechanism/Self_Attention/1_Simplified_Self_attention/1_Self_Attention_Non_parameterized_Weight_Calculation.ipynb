```python
import torch

# your journey starts with one step
inputs = torch.tensor([
    [0.43, 0.15, 0.89], # Your     (x^1)
    [0.55, 0.87, 0.66], # journey  (x^2)
    [0.57, 0.85, 0.64], # starts   (x^3)
    [0.22, 0.58, 0.33], # with     (x^4)
    [0.77, 0.25, 0.10], # one      (x^5)
    [0.05, 0.80, 0.55]  # step     (x^6)
] )
```

# For 2nd row [journey]

```python
query_2  = inputs[1]  
query_2
```

```python
attention_score_2 = torch.empty(inputs.shape[0])   # size = 6
attention_score_2
```

```python
for index, row_of_input in enumerate(inputs):
    attention_score_2[index] = torch.dot(row_of_input, query_2)
    
attention_score_2
```

```python
# Normalize 
attention_weight_2 = torch.softmax(attention_score_2, dim=-1)   # based on row 
attention_weight_2
```

# For all row

```python
attention_score = torch.empty(inputs.shape[0], inputs.shape[0])  # (6 * 6)
for i, row_of_input in enumerate(inputs):
    for j, row_as_query in enumerate(inputs):
        attention_score[i][j] = torch.dot(row_of_input, row_as_query)
    
attention_score
```

```python
attention_weight = torch.softmax(attention_score, dim=-1)   # based on col (sum of row = 0)
attention_weight
```

```python
print(torch.sum(attention_weight, dim=0))  # ❌
print(torch.sum(attention_weight, dim=-1))  # ✅  # based on col (sum of row = 0)
```

# Using Matrix Multiplication

```python
attention_score = inputs @ inputs.T
attention_score
```

```python
attention_weight = torch.softmax(attention_score, dim=-1)   # based on col (sum of row = 0)
attention_weight
```

```python
context_vector = attention_weight @ inputs   # (6,6) * (6,3) -> (6,3)
context_vector
```
